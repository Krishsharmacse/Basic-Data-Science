```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("titanic.csv")
```

```python
sns.histplot(x="Age",data=data)
plt.show()
```

```python
# it is greater than zero so it is right skewed
data["Age"].skew()
```
