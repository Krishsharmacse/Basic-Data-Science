```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import FunctionTransformer
```

```python
datset=pd.read_csv("loan.csv")
datset.head()
```

```python
transfarm=FunctionTransformer(func=np.log1p)
```

```python
transfarm.fit(datset["CoapplicantIncome"])
```

```python
datset["CoapplicantIncome"]=transfarm.transform(datset["CoapplicantIncome"])
```
