```python
import numpy as np
import pandas as pd
import scipy as sp
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("loan.csv")
dataset     
```

```python
dataset.describe()
```

```python
sns.boxplot(x="ApplicantIncome",data=dataset)
plt.show()
```

```python
len(dataset["ApplicantIncome"])
```

```python
q1=dataset["ApplicantIncome"].quantile(0.25)
q3=dataset["ApplicantIncome"].quantile(0.75)
app_iqr=q3-q1
lower_brack=q1-(1.5*app_iqr)
upper_brack=q3-(1.5*app_iqr)
print(lower_brack)
print(upper_brack)
```

```python
dataset=dataset[(dataset["ApplicantIncome"]<=upper_brack)& (dataset["ApplicantIncome"]>=lower_brack)]
```

```python
sns.boxplot(x="ApplicantIncome",data=dataset)
plt.show()
```
