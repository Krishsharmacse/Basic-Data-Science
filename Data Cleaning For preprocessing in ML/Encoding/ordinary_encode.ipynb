```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.DataFrame({"Size":["s","m","l","xl","s","l","m"]})
```

```python
ord_data=[["s","m","l","xl"]]
```

```python
oe=OrdinalEncoder(categories=ord_data)
```

```python
of=oe.fit(df[["Size"]])
of
```

```python
df["size_en"]=oe.transform(df[["Size"]])
```

```python
# By map function
ord1={"s":0,"m":1,"l":2,"xl":3}
```

```python
df["Size_map"]=df["Size"].map(ord1)
```
