```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
```

```python
dataset=pd.read_csv("loan.csv")
dataset.head()
```

```python
for col in ["Gender","Married"]:
    dataset[col].fillna(dataset[col].mode()[0],inplace=True)
```

```python
dataset.isnull().sum()
```

```python
data=["Gender","Married"]
```

```python
one_hot=OneHotEncoder(drop='first',sparse_output=False)
encoded=one_hot.fit_transform(dataset[data])
```

```python
encoded
```

```python
encoded_df=pd.DataFrame(encoded,columns=one_hot.get_feature_names_out(data))
```

```python
dataset=pd.concat([dataset.drop(["Gender","Married"],axis=1),encoded_df],axis=1)
```
